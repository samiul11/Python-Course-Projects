## Loading relevance judgments

```python
QRELS_FILE = "data/qrels.csv"

# TODO
```

```python
def load_qrels(file):
    gtruth = {}
    with open(file, "r") as fin:
        fin.readline()  # excluding header line from processing
        for line in fin:
            qid, did, rel = line.strip().split(",", 2)
            if not qid in gtruth:
                gtruth[qid] = {}
            gtruth[qid][did] = int(rel)
    return gtruth
qrels = load_qrels(QRELS_FILE)
```

## Computing NDCG scores

See Exercise #4 from [Lecture 8](https://github.com/kbalog/uis-dat640-fall2019/tree/master/exercises/lecture_08).

```python
def dcg(rel, p):
    dcg = rel[0]
    for i in range(1, min(p, len(rel))): 
        dcg += rel[i] / math.log(i + 1, 2)  # rank position is indexed from 1..
    return dcg
```

```python
sum_ndcg5 = 0
sum_ndcg10 = 0
sum_ndcg100 = 0

for qid, ranking in sorted(rankings.items()):
    gt = qrels[qid]    
    print("Query", qid)
    
    gains = [] # holds corresponding relevance levels for the ranked docs
    for doc_id in ranking: 
        gain = gt.get(doc_id, 0)
        gains.append(gain)
    print("\tGains:", gains)
    
    # relevance levels of the idealized ranking
    gain_ideal = sorted([v for _, v in gt.items()], reverse=True)
    print("\tIdeal gains:", gain_ideal)
    
    ndcg5 = dcg(gains, 5) / dcg(gain_ideal, 5)
    ndcg10 = dcg(gains, 10) / dcg(gain_ideal, 10)
    ndcg100 = dcg(gains, 100) / dcg(gain_ideal, 100)
    sum_ndcg5 += ndcg5
    sum_ndcg10 += ndcg10
    sum_ndcg100 += ndcg100
    
    print("\tNDCG@5:", round(ndcg5, 3), "\n\tNDCG@10:", round(ndcg10, 3),"\n\tNDCG@100:", round(ndcg100, 3))

print("Average")
print("\tNDCG@5:", round(sum_ndcg5 / len(rankings), 3), "\n\tNDCG@10:", round(sum_ndcg10 / len(rankings), 3),"\n\tNDCG@100:", round(sum_ndcg100 / len(rankings), 3))    
```
